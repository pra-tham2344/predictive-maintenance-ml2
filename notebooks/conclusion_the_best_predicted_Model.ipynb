```python
### ------------------------------------------------------------------------ ###
# --- Model Comparison & Sequence Length Analysis ---
### ------------------------------------------------------------------------ ###

from google.colab import drive
drive.mount('/content/drive')

# ------------------------------------------------- #
# --- 1. Imports ---
# ------------------------------------------------- #
import pandas as pd
import os
import matplotlib.pyplot as plt

# --- 2. Metrics for selected Models ---
# ------------------------------------------------------- #
# Metrics for XGBoost
rmse_xgb = 6.59
mae_xgb = 4.39
r2_xgb  = 0.97

# Metrics for Tuned XGB
rmse_tuned_xgb = 4.98
mae_tuned_xgb = 3.38
r2_tuned_xgb = 0.98
# Metrics for LSTM
rmse_lstm = 14.01
mae_lstm = 10.43
r2_lstm = 0.87

# Metrics for Random Forest
rmse_rf = 34.17
mae_rf = 24.39
r2_rf = 0.72

summary_df = pd.DataFrame({
  'Model' : ['XGBoost', 'Tuned XGB' , 'LSTM', 'RandomForest'],
  'RMSE' : [rmse_xgb, rmse_tuned_xgb, rmse_lstm, rmse_rf],
  'MAE' : [mae_xgb, mae_tuned_xgb, mae_lstm, mae_rf],
  'R2_Score' : [r2_xgb, r2_tuned_xgb, r2_lstm, r2_rf]
})

# --- 3. Save the comparison results ---
# ------------------------------------------------------------ #
comparison_path = "/content/drive/MyDrive/predictive_maintenance/code06_model_comparison.csv"
summary_df.to_csv(comparison_path, index=False)
print(f" Model comparison saved: {comparison_path}")

# --- Print table ---
print("\n--- Final Model Comparison ---")
print(summary_df)

# --- 4. Plot Model Comparison ( Set 'Model' as index) ---
# -------------------------------------------------------------- #
summary_df.set_index('Model')[['RMSE','MAE','R2_Score']].plot(kind='bar', figsize=(10,6))
plt.title("Model Performance Comparison")
plt.ylabel("Metrics Value")
plt.xticks(rotation=0)
plt.show()

# Conclusion
best_model = summary_df.loc[summary_df['RMSE'].idxmin(), 'Model']
print(f"\nConclusion: The best model for predicting RUL is {best_model}.")
print("Tuned XGBoost achieved the lowest RMSE and highest R², while being faster to train than LSTM/GRU.")
print("Therefore, Tuned XGBoost is recommended as the primary predictive model for this dataset.")


# --- 5. Save the data to drive ---
summary_df.to_csv("/content/drive/MyDrive/predictive_maintenance/model_comparison.csv", index=False)
plt.savefig("/content/drive/MyDrive/predictive_maintenance/rmse_mae_r2_comparison.png", dpi=300)
output_path = "/content/drive/MyDrive/predictive_maintenance/code06.ipynb"
print("Data Saved in drive")
print("Plots saved to drive")
print(f"Code 06 processed dataset saved to: {output_path}")

print("All plots and results for code 06 saved successfully!")


# ============================================================================= #
# Final Summary

# So in the final summary, we present the comparison of above mentioned 4 models
# and select the model which is having the least RMSE and highest R2_Score as compared to other 3 models.
# In our case, Tuned XGBoost Model is recommended as the Primary predictive model for this dataset.

# ============================================================================= #
```
